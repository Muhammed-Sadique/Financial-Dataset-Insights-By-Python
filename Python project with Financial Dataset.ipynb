```python
'Can you analyze and visualize the mean units sold for different countries using a bar plot ?'
```

```python
country_units_sold_mean = df.groupby('Country')['Units Sold'].mean()
country_units_sold_mean = country_units_sold_mean.reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Units Sold', data=country_units_sold_mean)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Mean Units Sold')
plt.title('Mean Units Sold by Country')
plt.show()
```
